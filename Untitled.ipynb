## Step 1: importing packages and dataset

```python
import tensorflow as tf
from tensorflow import keras
```

```python
fashion_mnist =  keras.datasets.fashion_mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()
```

```python
#check shape of the data 
print (X_train_full.shape)
print (X_test.shape)
```

## Step 2: Split train data to train and validation, and scale by dividing by 255

```python
X_val, X_train = X_train_full[:5000]/255.0, X_train_full[5000:]/255.0

y_val, y_train = y_train_full[:5000], y_train_full[5000:]
```

```python
#get the class names
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

## Step 3: Develop the sequential model

```python
from keras import Sequential
from keras.layers import Flatten, Dense

model = Sequential()
model.add(Flatten(input_shape=[28, 28])) #flatten the features to vonvert them to 1D array
model.add(Dense(300, activation = 'relu')) #add a dense layer with 300 neurons
model.add(Dense(100, activation = 'relu')) #add a dense layer with 100 neurons
model.add(Dense(10, activation = 'softmax')) #output layer with 10 outputs representing all the unique class labels

#compile the model
model.compile(optimizer='sgd', loss="sparse_categorical_crossentropy", metrics=['accuracy'])

#let's check the summary
model.summary()
```

```python
#fit the model

history = model.fit(X_train, y_train, epochs=30, validation_data=(X_val, y_val))
```

```python
#evaluate on the test set
model.evaluate(X_test, y_test)
```
